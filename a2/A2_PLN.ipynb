```python
import wget
import pandas as pd
import re
import nltk
```

Fazendo o download do dataset.

```python
url = "https://raw.githubusercontent.com/viniciusrpb/cic0269_natural_language_processing/main/corpus_tweets/twitter-2013train-A.txt"

file_name = wget.download(url)
print(file_name)
```

```python
with open(file_name) as file:
    dt = pd.DataFrame(file)
    print(dt)
```

Transformando o dataset

```python
dt1 = dt.apply(lambda row: row[0].split('\t', 2), axis=1, result_type='expand')
dt1.rename(columns={0: 'ID', 1: 'Sentiment', 2: 'Text'}, inplace=True)
dt1
```

### Preparando...

Baixando os pacotes para lemmatização e tokenização.

```python
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import RegexpTokenizer

tk = RegexpTokenizer(r'\w+|\$')
wn_lemmatizer = WordNetLemmatizer()
```

Função para padronizar e lemmatizar

```python
def get_standard_form(text):

    text = re.sub(r'\b(rofl|lol)\b', 'laugh', text)
    text = re.sub(r'\bthx\b', 'thanks', text)
    text = re.sub(r'\br\b', 'are', text)
    text = re.sub(r'\bu\b', 'you', text)
    text = re.sub(r':\)|:-\)|:D', 'happy', text)
    text = re.sub(r'(:\(|:-\()', 'sad', text)

    text = re.sub(r'http:.+/.+?(\s|\b)', ' ', text) # Removendo links
    # re.sub(r'\b\._\.\b', 'sad', text)

    return text


def lemmatize(text):

    res = text.split()
    res = [wn_lemmatizer.lemmatize(w, pos='v') for w in res]
    res = [wn_lemmatizer.lemmatize(w, pos='a') for w in res]

    return ' '.join(res)
```

Obtendo lista de stopwords em inglês.

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

sw = stopwords.words('english')
print(sw)
```

Preparando o texto a ser processado

```python
dt1['pre_proc_text'] = dt1['Text']
dt1
```

### Finalmente Executando
Removendo RT's

```python
dt1['pre_proc_text'] = dt1['pre_proc_text'].apply(lambda text: re.sub(r'\b(rt|RT)\b', '', text))
dt1
```

Padronização, lemmatização, além de outras transformações

```python
def resume(text):
    text = get_standard_form(text)
    text = text.lower()
    text = ' '.join(tk.tokenize(text)) # Limpando a pontuação
    text = lemmatize(text)

    return text

dt1['pre_proc_text'] = dt1['pre_proc_text'].apply(resume)
dt1
```

Limpando stopwords, e obtendo o dataset final


```python
dt1['pre_proc_text'] = dt1['pre_proc_text'].apply(lambda text: ' '.join([word for word in text.split() if word not in sw]))
dt1
```

```python
dt1.drop(columns='Text')
```

Obtendo as palavras dos tweets

```python
vocabulary = set()
for text in dt1['pre_proc_text']:
    vocabulary.update(text.split())

(vocabulary)
```
